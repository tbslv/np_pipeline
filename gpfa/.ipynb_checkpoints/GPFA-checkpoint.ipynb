```python
import funs.util as util
import funs.engine as engine
import funs.datamanager as datman
import matplotlib.pyplot as plt
import numpy as np

from neo_access import *
from neo.io import PickleIO
import pandas as pd
import os
```

```python
wd = 'D:/Work/Recordings/DATA/Neuropixel/190628/2019-06-28_09-27-20/SpikeSortingResults'

trials = 20
sweeplength = 22
samplerate = 30000



os.chdir(wd)
io = PickleIO(filename="vpl_block_20190628.pkl")
blk2 = io.read()[0]
dataframe = pd.read_pickle(os.path.join(wd,'vpl_block_df_20190628.pkl'))


expID = '002_'
sweep_Id = 0.0

selection_tmp = dataframe[(dataframe['expID'] == expID)& (dataframe['sweep_Id'] == sweep_Id)]
data = get_all_data(blk2,selection_tmp,raster=True,timestamps=False,spikes=False,aisignal=False)

neurons_num = len(data[0])
trials_num = 20
signal_len = int(selection_tmp['sweeplength'][0]*30000)
time = np.linspace(0, 20, signal_len)



# bin data

binsize= 0.1*30000
bins = np.arange(0,signal_len,binsize)
time_bin = np.linspace(0, 20, bins.size-1)
dat = np.zeros((neurons_num,trials_num,bins.size-1))

print('Data Dimensions are: {} x {} x {}'.format(neurons_num, trials_num, bins.size-1))

for clu in range(neurons_num):
    clu = int(clu)
    for tr in range(trials_num):
        tr = int(tr)
        
        x,y = np.histogram((data[0][clu][:,0][(data[0][clu][:,1]== tr)]).astype(int),bins=bins)
        dat[clu,tr,:] = x
```

```python
dat = dat[:,:,80:110]


training_set = datman.readNumpy(dat,0,3)
xdim = 3
maxEMiter = 100


# Initialize parameters using Poisson-PCA
initParams = util.initializeParams(xdim, training_set.ydim, training_set)
print('init params done ')
# Fit using vanilla (batch) EM
fitBatch = engine.PPGPFAfit(
	experiment 		= training_set,
	initParams 		= initParams,
	inferenceMethod = 'laplace',
	EMmode 			= 'Batch',
	maxEMiter 		= maxEMiter)
print('vanilla done ')
# Fit using online EM
'''fitOnline = engine.PPGPFAfit(
 	experiment 		= training_set,
 	initParams 		= initParams,
 	EMmode 			= 'Online',
 	maxEMiter 		= maxEMiter,
 	inferenceMethod = 'laplace',
	batchSize 		= 5)'''
print('online EM done ')
# Make plots
fitBatch.plotTrajectory();plt.show()
fitBatch.plotParamSeq()

#fitOnline.plotParamSeq();plt.show()


#fitOnline.plotTrajectory();plt.show()
```
